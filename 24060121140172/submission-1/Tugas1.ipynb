```python
# Nama : Fadhail Athaillah Bima
# Nim : 24060121140172
# Nama Dataset : Spambase
# Link Dataset : https://archive.ics.uci.edu/dataset/94/spambase


url = "http://archive.ics.uci.edu/ml/machine-learning-databases/spambase/spambase.data"
```

```python
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
```

```python
import pandas
# meng import library pandas agar dapat membaca data dari format CSV
```

```python
dataset = pandas.read_csv(url, names=names)
# digunakan untuk membaca data dari suatu file CSV yang diakses melalui URL dan menyimpannya dalam bentuk struktur data yang disebut "DataFrame"
```

```python
print(dataset.shape)
# digunakan untuk mencetak bentuk (shape) dari DataFrame dataset. Hasil cetakan ini akan menampilkan jumlah baris dan kolom dalam DataFrame.
```

```python
print(dataset.head(20))
# digunakan untuk mencetak 20 baris pertama (atau jumlah baris yang ditentukan) dari DataFrame dataset. Hasil cetakan ini akan menampilkan beberapa baris awal data dari DataFrame.
```

```python
print(dataset.groupby('class').size())
# digunakan untuk mengelompokkan data dalam DataFrame dataset berdasarkan kolom 'class' dan menghitung jumlah data dalam setiap kelompok. Hasil cetakan ini akan menampilkan jumlah data dalam setiap kelompok 'class' atau kategori yang ada di dalam kolom 'class'.
```

```python
print(dataset.describe())
# digunakan untuk menghasilkan ringkasan statistik dari dataset yang ada dalam objek DataFrame. Hasil cetakan ini akan berisi informasi statistik tentang setiap kolom numerik dalam dataset. 
```

```python
import matplotlib.pyplot as plt
dataset.plot(kind='box', subplots=True,layout=(3,2), sharex=False, sharey=False)
plt.show()
# digunakan untuk membuat diagram kotak (box plot) dari dataset menggunakan pustaka matplotlib. 
```

```python
dataset.hist()
plt.show()
# digunakan untuk membuat histogram dari dataset menggunakan pustaka matplotlib. 
```

```python
import pandas
from pandas.plotting import scatter_matrix
scatter_matrix(dataset)
plt.show()
# digunakan untuk membuat matriks sebaran (scatter matrix) dari dataset menggunakan pustaka pandas.
```
